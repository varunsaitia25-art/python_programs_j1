````python

year = int(input("Enter a year: "))

# Check conditions for leap year
if (year % 400 == 0):
    print(year, "is a Leap Year ✅ (divisible by 400)")
elif (year % 100 == 0):
    print(year, "is NOT a Leap Year ❌ (divisible by 100 but not by 400)")
elif (year % 4 == 0):
    print(year, "is a Leap Year ✅ (divisible by 4 but not by 100)")
else:
    print(year, "is NOT a Leap Year ❌ (not divisible by 4)")
````

````python
# Simple Login System using if, elif, and else

# Predefined correct username and password
correct_username = "admin"
correct_password = "12345"

# Ask the user to enter their credentials
username = input("Enter username: ")
password = input("Enter password: ")

# Check login conditions using if, elif, else
if username == correct_username and password == correct_password:
    print("✅ Login successful! Welcome,", username)

elif username != correct_username and password == correct_password:
    print("❌ Invalid username. Please try again.")

elif username == correct_username and password != correct_password:
    print("❌ Invalid password. Please try again.")


else:
    print("❌ Both username and password are incorrect!")
````
