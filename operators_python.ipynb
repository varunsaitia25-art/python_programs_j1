```python
#comparision operators- >,<,>=,<=,==,!= (return true or false)
a=5
b=10
c=10
d=20
print(a>b)
print(b<c)
print(b>=c)
print(b==c)
print(b!=d)
```

```python
# logical operators - and,or,not,
# and - this operators checking all conditions are true result is true rest case failed 
# or - this oiperators check either one conditions true true then result true rest cases failed 
a=10
b=20
c=20
d=60
print(a>b and b==c)
print(a<b and c<d)
print(a>b or b==c)
print(a>b and b!=c)
print (not a>b or b!=c)
```

```python
# membership operators - checking members are present in that sequence or collection - in , not in
names=["shiva","vikas","divya","alok"]
print("alok" in names)
print("chandra" in names)
print("shiva" not in names)
print("vikas" in names)
```

```python
# identity operators - memory identification of values - is , is not 
a=10000
b=10000
print(id(a))
print(id(a))
print(a is b)
```
